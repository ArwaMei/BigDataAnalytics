```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 3Ô∏è‚É£ Fetch the content of a web page
url = "https://books.toscrape.com/"
response = requests.get(url)

# Check the status of the response
if response.status_code == 200:
    print("‚úÖ Page fetched successfully!")
else:
    print("‚ùå Failed to fetch page:", response.status_code)
```

```python
# 4Ô∏è‚É£ Display the HTML source code
html_content = response.text
print("\n--- HTML Source Code (first 500 characters) ---\n")
print(html_content[:500])  # print only first 500 chars to avoid overflow
```

```python
# 5Ô∏è‚É£ Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")
```

```python
# 6Ô∏è‚É£ Extract all book titles and prices
books = soup.find_all("article", class_="product_pod")

book_data = []
for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    book_data.append({"Title": title, "Price": price})

# Show extracted data
print("\n--- Extracted Book Data ---")
for b in book_data[:5]:  # show first 5 books
    print(b)
```

```python
df = pd.DataFrame(book_data)
df.to_csv("books_data.csv", index=False, encoding="utf-8")
print("\nüíæ Data saved to 'books_data.csv' successfully!")
```
